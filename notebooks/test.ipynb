```python
import sumolib

trips_file ='../sumo_config/processed_trips_old.xml'
veh_trips = sumolib.output.parse_sax__asList(trips_file, "trip", ['id','depart'])
bus_trips = sumolib.output.parse_sax__asList(trips_file, "vehicle", ['id','depart'])
```

```python
times = []

for trip in veh_trips:
    times.append(trip['depart'])

for trip in bus_trips:
    times.append(trip['depart'])
```

```python
import matplotlib.pyplot as plt

plt.hist(times, bins=24*4)
```
